```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10,5))
plt.grid()
plt.show()
```

```python
x=[1,2,3,4,5,6,7]
temp=[32,36,38,30,39,40,35]
plt.plot(x,temp)
plt.grid()
plt.show()
```

```python
x=[1,2,3,4,5,6,7]
temp=[32,36,38,30,39,40,35]
plt.plot(x,temp)
plt.xlabel("Day")
plt.ylabel("Temp")
plt.title("Climat")
plt.grid()
plt.show()
```

```python
x=[1,2,3,4,5,6,7]
temp=[32,36,38,30,39,40,35]
plt.plot(x,temp, marker="o", color="red", linestyle="dashed")
plt.xlabel("Day")
plt.ylabel("Temp")
plt.title("Climate")
plt.grid()
plt.show()
```

```python
company=["Relience", "Adani", "Tata"]
revenue=[110,120,90]
xpos=np.arange(len(company))
plt.bar(xpos,revenue)
```

```python
company=["Relience", "Adani", "Tata"]
revenue=[110,120,90]
xpos=np.arange(len(company))
plt.xticks(xpos,company)
plt.bar(xpos,revenue,width=0.2,color="green",label="Revenue")
plt.legend()
plt.show()
```

```python
x=["USA", "China", "Japan", "Australia"]
G=[40,40,20,18]
S=[44,27,12,19]
B=[42,24,13,16]
b_bronze=list(np.add(G,S))
plt.bar(x,G,width=0.4,label="Gold",color="yellow")
plt.bar(x,S,0.4,bottom=G,label="Silver",color="Gray")
plt.bar(x,S,0.4,bottom=b_bronze,label="Bronze",color="Red")
plt.show()
```
